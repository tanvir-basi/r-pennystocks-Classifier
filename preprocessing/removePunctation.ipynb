```python
import string
import mysql.connector
import re
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="",
  database="Finance Project",
)
mycursor = mydb.cursor()
mycursor.execute("SHOW DATABASES")

for x in mycursor:
  print(x)

mycursor.execute("SHOW TABLES")

for x in mycursor:
  print(x)
```

```python
mycursor.execute("SELECT selftext,title FROM final_table")

myresult = mycursor.fetchall()

for row in myresult:
    text = row[0]
    title = row[1]

    # remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # remove numbers
    text = re.sub(r'\d+', '', text)
    # remove double spaces
    text = re.sub(r'\s+', ' ', text)
    #  remove newlines
    text = text.replace('\n', '')
    # remove tabs
    text = text.replace('\t', '')
    #  remove links
    text = re.sub(r'http\S+', '', text)

    mycursor.execute("UPDATE final_table SET selftext = %s WHERE title = %s", (text, title))
    mydb.commit()
    print(mycursor.rowcount, "record(s) affected")
```
