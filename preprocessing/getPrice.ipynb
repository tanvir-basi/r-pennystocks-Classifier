```python
from lib2to3.pgen2 import token


mycursor.execute("SELECT stock,created_utc,title FROM pennystocks")

myresult = mycursor.fetchall()

limit = 1

for stock in myresult:
    while True:
        try:
            ticker = stock[0]
            price_today = Stock(ticker).get_quote().latestPrice[ticker]
            utc = stock[1]
            start = time.strftime('%Y-%m-%d', time.localtime(utc))
            price_post = get_historical_data(ticker, start, close_only=True).close[0]
            print(ticker, 'ticker')
            print(price_today, 'price_today')
            print(price_post, 'price_ticker_date')
            print(start, 'post date')
            # percentage change
            change = (price_today - price_post) / price_post
            change_clean = str(round(change, 2))
            print(change_clean, 'change')
            print('\n')
            # insert change_clean into database
            mycursor.execute("UPDATE pennystocks SET price_change = %s WHERE title = %s", (change_clean, stock[2]))
            break
        except (iexfinance.utils.exceptions.IEXQueryError, KeyError, ValueError, AttributeError, TypeError, IndexError, ZeroDivisionError) as e:
            print("Error:")
            print('\n')
            break

    # limit -= 1
    if limit == 0:
        break

mydb.commit()
print(mycursor.rowcount, "record(s) affected")
```
